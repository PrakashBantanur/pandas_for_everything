#### Statistical Graphics using matplotlib

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
print(tips.info())
```

### Univariate
#### Histograms

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histograms of Total bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
# fig.show()
```

### Bivariate
#### Scatter plots

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total bill vs. Tips')
axes1.set_ylabel('Tip')
axes1.set_xlabel('Total Bill')
```

#### Boxplot

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'], tips[tips['sex'] == 'Male']['tip']], labels = ['Female', 'Male'])
axes1.set_title('Boxplot of Tips by sex')
axes1.set_ylabel('Tip')
axes1.set_xlabel('Sex')
# boxplot.show()
```

#### Multivariable data

```python
def decode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
    
tips['sex_color'] = tips['sex'].apply(decode_sex)

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(x = tips['total_bill'], y = tips['tip'], s = tips['size']*10, c = tips['sex_color'], alpha = 0.5)
axes1.set_title('Total bill Vs. Tip colored by sex and sized by size ')
axes1.set_ylabel('Tip')
axes1.set_xlabel('Total Bill')
```

# Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

#### Univariate

```python
hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histograms with Destiny plot')
```

```python
hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)
ax.set_title('Total Bill Histograms with Destiny plot')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)
ax.set_title('Total Bill Destiny plot')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Total Bill with Destiny and Rug plot')
ax.set_xlabel('Total Bill')
```

#### Count plot

```python
count, ax = plt.subplots()
ax = sns.countplot('day', data = tips)
ax.set_title('Count of Days')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Frequency')
```

```python
tips.day.value_counts()
```

### Bivariate Data
#### Scatter plot

```python
scatter, ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips) # fit_reg=False
ax.set_title('Scatterplot of Total Bill and tips')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
fig = sns.lmplot(x = 'total_bill', y = 'tip', data=tips)
```

#### Hexbin Plot

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data = tips, kind='hex')  #You can remove( kind='hex') and check
hexbin.set_axis_labels(xlabel='total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint plot of Total bill and Tip', fontsize=10, y=1.03)
```

#### 2D Density plot

```python
kde, ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'], data2=tips['tip'], shade=True) # Try (shade=False)
ax.set_title('Kernel density plot of Total Bill and tips')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
kde_joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

#### Bar Plot

```python
barplot, ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average bill by time of day')
ax.set_ylabel('Average total Bill')
ax.set_xlabel('Time of day')
```

#### Violin Plot

```python
violin, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violinplot of Total Bill by time of day')
ax.set_ylabel('Total Bill')
ax.set_xlabel('Time of day')
```

#### Boxplot

```python
boxplot, ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of Total Bill by time of day')
ax.set_ylabel('Total Bill')
ax.set_xlabel('Time of day')
```

```python
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)
fig.tight_layout()
```
